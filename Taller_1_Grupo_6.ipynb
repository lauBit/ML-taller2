```python
# Importación de bibliotecas
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules, fpgrowth, fpmax
from mlxtend.preprocessing import TransactionEncoder
from surprise import Dataset, Reader, KNNBasic, KNNBaseline
from surprise.model_selection import train_test_split
from surprise import accuracy
from surprise.model_selection import KFold
from surprise import Dataset, Reader
```

```python
# Carga del conjunto de datos
df= pd.read_csv('/content/steam-200k.csv')
```

```python
# Visualización de las primeras 10 filas del DataFrame
df.head(10)
```

```python
# Renombrar columnas del DataFrame
df.columns = ['user_id', 'game-title', 'behavior-name', 'value','valor_0']
```

```python
df
```

```python
# Crear una tabla dinámica para reorganizar los datos
df_pivoted = df.pivot_table(index=['user_id', 'game-title'],
                            columns='behavior-name',
                            values='value',
                            aggfunc='first').reset_index()
```

```python
# Eliminar el nombre de las columnas
df_pivoted.columns.name = None

# Renombrar columnas específicas
df_pivoted.rename(columns={'purchase': 'purchase_value', 'play': 'play_time'}, inplace=True)
```

```python
# Visualización de las primeras 10 filas del DataFrame pivotado
df_pivoted.head(10)
```

```python
# Descripción del DataFrame pivotado
df_pivoted.describe()
```

```python
# Información del DataFrame pivotado
df_pivoted.info()
```

```python
# Filtrar y depurar el DataFrame
df_pivoted_depurado = df_pivoted[['user_id', 'game-title', 'play_time']].dropna()
```

```python
# Ajustar los valores de tiempo de juego ajustando los datos logaritmicamente
df_pivoted_depurado['play_time_ajustado'] = df_pivoted_depurado['play_time'].apply(lambda x: np.log1p(x) if x > 0 else 0)
```

```python
df_pivoted_depurado.describe()
```

```python
# Crear un DataFrame limpio con las columnas necesarias
df_limpio = df_pivoted_depurado[['user_id', 'game-title', 'play_time_ajustado']]
```

```python
# Definir el rango de calificación
reader = Reader(rating_scale=(df_limpio['play_time_ajustado'].min(), df_limpio['play_time_ajustado'].max()))
```

```python
# Cargar los datos en el formato adecuado
data = Dataset.load_from_df(df_limpio, reader)
```

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
trainset, testset = train_test_split(data, test_size=0.25, random_state=123)
```

```python
print(trainset.n_items)
```

```python
# Inicializar el modelo KNN
knn = KNNBasic()
```

```python
# Ajustar el modelo KNN al conjunto de entrenamiento
knn.fit(trainset)
```

```python
# Opciones de similitud del modelo KNN
knn.sim_options
```

```python
# Predecir sobre el conjunto de prueba
predictions = knn.test(testset)
```

```python
# Visualizar las predicciones
predictions
```

```python
# Calcular y mostrar el RMSE de las predicciones
accuracy.rmse(predictions)
```

## **Algoritmo SVD**

```python
# Algoritmo SVD para comparación
from surprise import SVD
from surprise.model_selection import cross_validate
```

```python
# Inicializar el modelo SVD
svd = SVD()

# Ajustar el modelo SVD al conjunto de entrenamiento
svd.fit(trainset)
```

```python
# Validación cruzada del modelo SVD
results = cross_validate(svd, data, measures=['RMSE', 'MAE'], cv=5)
```

```python
# Imprimir resultados de RMSE y MAE
print('RMSE:', results['test_rmse'].mean())
print('MAE:', results['test_mae'].mean())
```

```python
# Predecir sobre el conjunto de prueba con el modelo SVD
predictions_1 = svd.test(testset)
```

```python
# Calcular y mostrar el RMSE de las predicciones SVD
accuracy.rmse(predictions_1)
```
